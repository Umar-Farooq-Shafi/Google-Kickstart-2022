Flood fill

```python
import sys
print(sys.getrecursionlimit())

DIRECTIONS: tuple = ((0, 1), (-1, 1), (-1, 0), (0, -1), (1, -1), (1, 0))


def flood(board, x, y, color):
    board[x][y] = color.lower()
    for dx, dy in DIRECTIONS:
        if 0 <= x + dx < len(board) and 0 <= y + dy < len(board):
            if board[x + dx][y + dy] == color:
                flood(board, x + dx, y + dy, color)


def check_winners(board, n):
    for i in range(n):
        if board[i][0] == "B":
            flood(board, i, 0, "B")
        if board[0][i] == "R":
            flood(board, 0, i, "R")

    for i in range(n):
        if board[i][n - 1] == "b":
            return "B"
        if board[n - 1][i] == "r":
            return "R"

    return "."


def count_stones(board):
    num_red, num_blue = 0, 0
    for row in board:
        for col in row:
            if col == "R":
                num_red += 1
            elif col == "B":
                num_blue += 1
    return num_red, num_blue

def deepcopy(board):
    return [row[:] for row in board]


def solve(original_board, n):
    num_red, num_blue = count_stones(original_board)
    if abs(num_red - num_blue) > 1:
        return "Impossible"

    board = deepcopy(original_board)
    winner = check_winners(board, n)
    if winner == ".":
        return "Nobody wins"

    if (winner == "B" and num_blue < num_red) or (winner == "R" and num_red < num_blue):
        return "Impossible"

    for row in range(n):
        for col in range(n):
            if original_board[row][col] == winner:
                board = deepcopy(original_board)
                board[row][col] = "."
                if check_winners(board, n) == ".":
                    if winner == "B":
                        return "Blue wins"
                    elif winner == "R":
                        return "Red wins"

    return "Impossible"

if __name__ == "__main__":
    if line := input.split('\n')[1:-1]:
    
        # test_cases = int(input())
        index = 0
        test_cases = int(str(line[index]).strip())
        index += 1
        for test_case in range(1, test_cases + 1, 1):
            # board_size = int(input())
            board_size = int(str(line[index]).strip())
            index += 1
            board = []
            for _ in range(board_size):
                # board.append(list(input().strip()))
                board.append(list(str(line[index]).strip()))
                index += 1

            ans = solve(board, board_size)

            print("Case #{}: {}".format(test_case, ans))

```

```python
def gen_next_hex(left, right):
    right_dir = (right[0] - left[0], right[1] - left[1])
    
    for index, dir in enumerate(DIRECTIONS):
        if right_dir == dir:
            next_dir = DIRECTIONS[(index + 1) % len(DIRECTIONS)]
            return (left[0] + next_dir[0], left[1] + next_dir[1])
        
    raise Exception
```

```python
def step(padded_path, colour, left, right):
    next_hex = gen_next_hex(left, right)
    if padded_path[next_hex[0]][next_hex[1]] == colour:
        return next_hex, right
    else:
        return left, next_hex
```

```python
def blue_path_south(padded_board, m):
    # start at the southest corner with a blue stone on the left and 
    # a red stone on the right
    left, right = (m - 1, 0), (m - 1, 1)
    path = set()
    
    # follow a path between the hex, always keeping blue stone on the left,
    while left[1] < m - 1: # stop when reach the east side
        path.add(left)
        _, right = step(padded_board, 'B', left, right)
        if right[0] == 0:
            return None   # break, if we reach the north side
    
    return path
```
